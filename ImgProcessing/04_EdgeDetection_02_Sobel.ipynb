## Edge Detection: Sobel

```python
# Computes convolution between two matrix X and F
def convolve_np(X, F):
    X_height,X_width = X.shape[0],X.shape[1]
    F_height,F_width = F.shape[0],F.shape[1]
    
    H = np.int64((F_height - 1) / 2)
    W = np.int64((F_width - 1) / 2)
    
    out = np.zeros((X_height, X_width))
    
    for i in np.arange(H, X_height-H):
        for j in np.arange(W, X_width-W):
            sum = 0
            for k in np.arange(-H, H+1):
                for l in np.arange(-W, W+1):
                    a = X[i+k, j+l]
                    w = F[H+k, W+l]
                    sum += (w * a)
            out[i,j] = sum
    return out
```

### Sobel Horizontal Gradient Operator detects Vertical Edge

#### Sobel operated image with absolute values

```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
import time
img = cv2.imread('images/butterfly.jpg', cv2.IMREAD_GRAYSCALE)
print(img.dtype)
height = img.shape[0]
width = img.shape[1]
print('height=',height,'width = ',width)

Hx = np.array([[1, 0, -1],
               [2, 0, -2],
               [1, 0, -1]],)
img_x_Edge = np.absolute(convolve_np(img, Hx))

img_x_Scaled = 255*(img_x_Edge - img_x_Edge.min())/ (img_x_Edge.max() - img_x_Edge.min())
cv2.imwrite('images/sobel_Hx.jpg', img_x_Scaled)

plt.imshow(img_x_Scaled, cmap = 'gray', interpolation = 'bicubic')
plt.xticks([]), plt.yticks([])  # to hide tick values on X and Y axis
plt.show()
```

### Sobel Vertical Gradient Operator detects Horizontal Edges

#### Sobel operated image with absolute values

```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
import time
img = cv2.imread('images/butterfly.jpg', cv2.IMREAD_GRAYSCALE)
print(img.dtype)
height = img.shape[0]
width = img.shape[1]
print('height=',height,'width = ',width)

Hy = np.array([[1, 2, 1],
               [0, 0, 0],
               [-1, -2, -1]])
img_y_Edge = np.absolute(convolve_np(img, Hy))
img_y_Scaled= 255*(img_y_Edge - img_y_Edge.min())/ (img_y_Edge.max() - img_y_Edge.min())
cv2.imwrite('images/sobel_Hy.jpg', img_y_Scaled)

plt.imshow(img_y_Scaled, cmap = 'gray', interpolation = 'bicubic')
plt.xticks([]), plt.yticks([])  # to hide tick values on X and Y axis
plt.show()
```
