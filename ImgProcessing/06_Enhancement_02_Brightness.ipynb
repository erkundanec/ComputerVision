## Brightness

Increase the brightness of each pixel value

```python
import numpy as np
import cv2
import math
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('images/low_contrast_1.jpg', cv2.IMREAD_GRAYSCALE)
imgOrig = img.copy()
print(img.dtype)
height = img.shape[0]
width = img.shape[1]
print('height=',height,'width = ',width)
```

```python
brightness = 100
img = imgOrig.copy()
for i in np.arange(height):
    for j in np.arange(width):
        a = img.item(i,j)
        b = a + brightness
        if b > 255:
            b = 255
        img.itemset((i,j), b)

#cv2.imwrite('images/brightness.jpg', img)

plt.figure()
plt.imshow(imgOrig,'gray')
plt.figure()
plt.imshow(img,'gray')
```
