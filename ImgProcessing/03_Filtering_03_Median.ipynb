## Median Filter

```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

## Written from scratch

```python
img = cv2.imread('images/plane_noisy.png', cv2.IMREAD_GRAYSCALE)
img_out = img.copy()
print(img.dtype)
height = img.shape[0]
width = img.shape[1]
print('height=',height,'width = ',width)
```

```python
height = img.shape[0]
width = img.shape[1]

kernelSize = 5

centerPixel = int((kernelSize*kernelSize - 1) / 2)

for i in np.arange(kernelSize, height-kernelSize):
    for j in np.arange(kernelSize, width-kernelSize):
        neighbors = []
        for k in np.arange(-kernelSize, kernelSize+1):
            for l in np.arange(-kernelSize, kernelSize+1):
                a = img.item(i+k, j+l)
                neighbors.append(a)
        neighbors.sort()
        median = neighbors[centerPixel]
        b = median
        img_out.itemset((i,j), b)
```

```python
#cv2.imwrite('images/filter_median.jpg', img_out)

plt.figure()
plt.imshow(img,'gray')
plt.figure()
plt.imshow(img_out,'gray')
plt.show()
```
