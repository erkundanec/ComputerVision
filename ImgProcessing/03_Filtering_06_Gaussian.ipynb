## Gaussian Filter

```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('images/low_contrast_1.jpg', cv2.IMREAD_GRAYSCALE)
img_out = img.copy()
print(img.dtype)
height = img.shape[0]
width = img.shape[1]
print('height=',height,'width = ',width)
```

```python
height = img.shape[0]
width = img.shape[1]

gauss = (1.0/57) * np.array(
        [[0, 1, 2, 1, 0],
        [1, 3, 5, 3, 1],
        [2, 5, 9, 5, 2],
        [1, 3, 5, 3, 1],
        [0, 1, 2, 1, 0]])

gauss.sum()
```

```python
for i in np.arange(2, height-2):
    for j in np.arange(2, width-2):        
        sum = 0
        for k in np.arange(-2, 3):
            for l in np.arange(-2, 3):
                a = img.item(i+k, j+l)
                p = gauss[2+k, 2+l]
                sum = sum + (p * a)
        b = sum
        img_out.itemset((i,j), b)
```

```python
#cv2.imwrite('images/filter_gauss.jpg', img_out)

plt.figure()
plt.imshow(img,'gray')
plt.figure()
plt.imshow(img_out,'gray')
plt.show()
```
