## Auto Contrast

```python
import numpy as np
import cv2
import math
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('images/low_contrast_1.jpg', cv2.IMREAD_GRAYSCALE)
imgOrig = img.copy()
print(img.dtype)
height = img.shape[0]
width = img.shape[1]
print('height=',height,'width = ',width)
```

```python
min = 255
max = 0
for i in np.arange(height):
    for j in np.arange(width):
        a = img.item(i,j)
        if a > max:
            max = a
        if a < min:
            min = a
            
print('min value = ', min)
print('max value = ', max)
            
for i in np.arange(height):
    for j in np.arange(width):
        a = img.item(i,j)
        b = float(a - min) / (max - min) * 255
        img.itemset((i,j), b)

#cv2.imwrite('images/auto_contrast.jpg', img)

plt.figure()
plt.imshow(imgOrig,'gray')
plt.axis("off")
plt.figure()
plt.imshow(img,'gray')
plt.axis("off")
```

```python
print('Before auto contrast: ',imgOrig.min(),imgOrig.max())
print('After auto contrast: ',img.min(),img.max())
```
