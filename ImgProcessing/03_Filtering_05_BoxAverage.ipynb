## Box Filter

```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('images/low_contrast_1.jpg', cv2.IMREAD_GRAYSCALE)
print(img.dtype)
height = img.shape[0]
width = img.shape[1]
print('height=',height,'width = ',width)
```

```python
img_out = img.copy()

height = img.shape[0]
width = img.shape[1]

kernelSize = 5
temp = np.int((kernelSize-1)/2)

for i in np.arange(temp, height-temp):
    for j in np.arange(temp, width-temp):        
        sum = 0
        for k in np.arange(-temp, temp+1):
            for l in np.arange(-temp, temp+1):
                a = img.item(i+k, j+l)
                sum = sum + a
        b = int(sum / (kernelSize*kernelSize))
        img_out.itemset((i,j), b)

#cv2.imwrite('images/filter_box.jpg', img_out)
plt.figure()
plt.imshow(img,'gray')
plt.figure()
plt.imshow(img_out,'gray')
plt.show()
```
